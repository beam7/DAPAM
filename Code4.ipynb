## Primary Bar Graphs

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

input_csv1 = "/content/drive/My Drive/Chen Lab/Antimicrobial Peptide Project/filtered_antibacterial_for_comparison.csv"
input_csv2 = "/content/drive/My Drive/Chen Lab/Antimicrobial Peptide Project/processed_annotatedcopy_peptide_data_updated.csv"

# === Load datasets ===
df1 = pd.read_csv(input_csv1)
df2 = pd.read_csv(input_csv2)

# === Define features to include in main figure ===
features = [
    "Length (number of amino acids)",
    "Net Charge",
    "Aromaticity",
    "Percent C"
]

# === Create subplot grid: 2 rows (1 per feature), 4 columns (Peptipedia | DAPAM) ===
fig, axes = plt.subplots(2, 4, figsize=(27, 14))
axes = axes.flatten()

for i, feature in enumerate(features):
    data1 = df1[feature].dropna()
    data2 = df2[feature].dropna()

    # Compute consistent x-axis scale and reduce number of bins
    combined = np.concatenate([data1, data2])
    bin_count = 18
    #bin_count = max(5, len(np.histogram_bin_edges(combined, bins='auto')) // 4)
    bins = np.histogram_bin_edges(combined, bins=bin_count)

    # Plot Peptipedia
    ax1 = axes[i * 2]
    ax1.hist(data1, bins=bins, color='blue', edgecolor='black', alpha=0.7)
    ax1.set_title(f"Peptipedia", fontsize=24)
    ax1.set_xlabel(feature, fontsize=24)
    ax1.set_ylabel("Frequency", fontsize=24)
    ax1.tick_params(axis='x', labelsize=24)
    ax1.tick_params(axis='y', labelsize=24)

    # Plot DAPAM
    ax2 = axes[i * 2 + 1]
    ax2.hist(data2, bins=bins, color='red', edgecolor='black', alpha=0.7)
    ax2.set_title(f"DAPAM", fontsize=24)
    ax2.set_xlabel(feature, fontsize=24)
    ax2.set_ylabel("Frequency", fontsize=24)
    ax2.tick_params(axis='x', labelsize=24)
    ax2.tick_params(axis='y', labelsize=24)

# === Final formatting ===
plt.tight_layout()
plt.subplots_adjust(hspace=0.6)  # Increase vertical space between rows
plt.savefig("/content/drive/My Drive/Chen Lab/Antimicrobial Peptide Project/Figure_Main_Comparison_Histograms.png", dpi=300)
plt.show()
```

## Heat Map of Cohen's d-values

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import normaltest, mannwhitneyu, ttest_ind, ks_2samp

# Load your CSVs
input_csv1 = "/content/drive/My Drive/Chen Lab/Antimicrobial Peptide Project/filtered_antibacterial_for_comparison.csv"
input_csv2 = "/content/drive/My Drive/Chen Lab/Antimicrobial Peptide Project/processed_annotatedcopy_peptide_data_updated.csv"

df1 = pd.read_csv(input_csv1)
df2 = pd.read_csv(input_csv2)

numeric_features = [
    "Length (number of amino acids)", "Molecular Weight (Daltons)", "Net Charge", "Charge",
    "Charge Density (Charge/Dalton)", "Instability Index", "Aromaticity", "Aliphatic Index",
    "Boman Index (kcal/mol)", "Isoelectric Point (pH)", "Hydrophobic Ratio",
    "Percent A", "Percent R", "Percent N", "Percent D", "Percent C", "Percent E", "Percent Q",
    "Percent G", "Percent H", "Percent I", "Percent L", "Percent K", "Percent M", "Percent F",
    "Percent P", "Percent S", "Percent T", "Percent W", "Percent Y", "Percent V"
]

df1 = df1[numeric_features].dropna()
df2 = df2[numeric_features].dropna()

# Function to compute Cohen's d
def cohens_d(x, y):
    return (np.mean(x) - np.mean(y)) / np.sqrt((np.var(x, ddof=1) + np.var(y, ddof=1)) / 2)

# Calculate Cohen's d for all features
cohen_d_vals = {}
for feature in numeric_features:
    d = cohens_d(df1[feature], df2[feature])
    cohen_d_vals[feature] = d

# Convert to DataFrame
cohen_df = pd.DataFrame.from_dict(cohen_d_vals, orient='index', columns=["Cohen's d"])
cohen_df = cohen_df.sort_values(by="Cohen's d")  # Optional sort

# Create annotation matrix with only |d| > 0.2
annot = cohen_df.copy().round(2).astype(str)
annot[cohen_df.abs() <= 0.2] = ""

# Plot heatmap
plt.figure(figsize=(14, 4))  # Less tall
sns.heatmap(
    cohen_df.T, cmap="coolwarm", center=0,
    annot=annot.T, fmt="", annot_kws={"size": 8},
    cbar_kws={"label": "Cohen's d"},
    linewidths=0.4
)

plt.title("Cohens d-values Between Peptipedia and DAPAM by Feature", fontsize=14)
plt.yticks(rotation=0, fontsize=12)
plt.xticks(rotation=30, ha='right', fontsize=10)
plt.tight_layout()
plt.savefig("/content/drive/My Drive/Chen Lab/Antimicrobial Peptide Project/cohens_d_heatmap_filtered.png", dpi=300)
plt.show()
```

## Volcano Plot

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import mannwhitneyu
from scipy.stats import normaltest

# Load datasets
df1 = pd.read_csv("/content/drive/My Drive/Chen Lab/Antimicrobial Peptide Project/filtered_antibacterial_for_comparison.csv")
df2 = pd.read_csv("/content/drive/My Drive/Chen Lab/Antimicrobial Peptide Project/processed_annotatedcopy_peptide_data_updated.csv")

# Numeric features
numeric_features = [
    "Length (number of amino acids)", "Molecular Weight (Daltons)", "Net Charge", "Charge",
    "Charge Density (Charge/Dalton)", "Instability Index", "Aromaticity", "Aliphatic Index",
    "Boman Index (kcal/mol)", "Isoelectric Point (pH)", "Hydrophobic Ratio", "Percent A", "Percent R",
    "Percent N", "Percent D", "Percent C", "Percent E", "Percent Q", "Percent G", "Percent H",
    "Percent I", "Percent L", "Percent K", "Percent M", "Percent F", "Percent P", "Percent S",
    "Percent T", "Percent W", "Percent Y", "Percent V"
]
'''
# Function to compute Cohen's d
def cohens_d(x, y):
    nx, ny = len(x), len(y)
    if nx < 2 or ny < 2:
        return np.nan
    pooled_std = np.sqrt(((nx - 1)*np.var(x, ddof=1) + (ny - 1)*np.var(y, ddof=1)) / (nx + ny - 2))
    return (np.mean(x) - np.mean(y)) / pooled_std if pooled_std else np.nan
'''
# Function to compute Cohen's d
def cohens_d(x, y):
    return (np.mean(x) - np.mean(y)) / np.sqrt((np.var(x, ddof=1) + np.var(y, ddof=1)) / 2)

# Run tests and calculate stats
results = []
for feature in numeric_features:
    x = df1[feature].dropna()
    y = df2[feature].dropna()

    if len(x) < 2 or len(y) < 2:
        continue

    # Mann-Whitney for p-value
    try:
        stat, p = mannwhitneyu(x, y, alternative='two-sided')
    except Exception:
        p = np.nan

    # Cohen's d
    d = cohens_d(x, y)

    results.append({
        "Feature": feature,
        "p_value": p,
        "Cohen_d": d,
        "-log10(p_value)": -np.log10(p) if p > 0 else np.nan
    })

# Convert to DataFrame
results_df = pd.DataFrame(results).dropna()

# Volcano Plot
plt.figure(figsize=(12, 6))
sns.scatterplot(
    data=results_df,
    x="Cohen_d",
    y="-log10(p_value)",
    hue=results_df["p_value"] < 0.05,
    palette={True: "red", False: "gray"}
)

# Add horizontal and vertical lines
plt.axvline(x=0.2, linestyle="--", color="green", label="Small effect (d=0.2)")
plt.axvline(x=-0.2, linestyle="--", color="green")
plt.axhline(y=-np.log10(0.05), linestyle="--", color="blue", label="p = 0.05")

# Annotate red dots
for _, row in results_df.iterrows():
    if row["p_value"] < 0.05:
        plt.text(row["Cohen_d"], row["-log10(p_value)"] + 0.15, row["Feature"], fontsize=10, ha='right')

plt.xlabel("Cohen's d", fontsize=12)
plt.ylabel("-log10(p-value)", fontsize=12)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.ylim(0, results_df["-log10(p_value)"].max() + 0.5)
plt.title("Feature Discrimination Between DAPAM and Peptipedia: Volcano Plot", fontsize = 14)
plt.legend()
plt.tight_layout()
plt.savefig("/content/drive/My Drive/Chen Lab/Antimicrobial Peptide Project/volcano_plot.png", dpi=300)
plt.show()
```
